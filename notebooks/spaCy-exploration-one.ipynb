Begin by importing spaCy, we will work with the cleaned data just to illustrate what we can do at a basic level

```python
import spacy
import numpy as np
import pandas as pd

df = pd.read_csv("../data/clean_technology_reviews.csv")
# choosing a review
a = df['reviewText'].values[8]
a = str(a)

nlp = spacy.load("en_core_web_lg")
import en_core_web_lg
nlp = en_core_web_lg.load()
doc = nlp(a)
print([(w.text, w.pos_) for w in doc])
```

```python
# we can display similarity of words, however this doesn't seem to work with words outside of the corpus
w1 = "I  you"
w2 = "you don't everyone"

# w1 = nlp.vocab[w1]
# w2 = nlp.vocab[w2]
w1 = nlp(w1)
w2 = nlp(w2)
```
