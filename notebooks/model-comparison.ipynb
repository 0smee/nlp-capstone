```python
import matplotlib.pyplot as plt
# Create a figure and an axis
fig, ax = plt.subplots()

# Set the limits of the graph
ax.set_xlim(0, 0.6)
ax.set_ylim(0, 1)

ax.set_xlabel('love', fontsize=14)
ax.set_ylabel('you', fontsize=14)

ax.quiver(0, 0, 0.4, 0.8, angles='xy', scale_units='xy', scale=1, color="navy")

ax.annotate('("love","you")=(0.4, 0.8)', xy=(0.4, 0.8), textcoords="offset points", xytext=(10,10), ha='center', fontsize=14)

plt.show()
```
